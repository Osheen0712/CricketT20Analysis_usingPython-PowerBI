### Loading the libraries

```python
import pandas as pd
import json
```

### Loading the json files in your notebook

### 1st :  MATCH RESULTS

```python
with open('cricket_t20_json_files/t20_wc_match_results.json') as f:
    data=json.load(f)

df_match = pd.DataFrame(data[0]['matchSummary'])

print(df_match.shape)
df_match.head(10)
```

```python
# Renaming column 'scorecard'
df_match.rename({'scorecard':'match_id'},axis=1,inplace=True)
df_match.head(2)
```

```python
# Creating a dictionary 'match_dict' and with keys as 'team1 Vs team2' as well as 'team2 Vs team1' and their values as the 'match_id' 

match_dict={}

for index,row in df_match.iterrows():
    key1 = row['team1'] + ' Vs ' + row['team2']
    key2 = row['team2'] + ' Vs ' + row['team1']
    
    match_dict[key1] = row["match_id"]
    match_dict[key2] = row["match_id"]
match_dict
```

```python
# Exporting the file
df_match.to_csv('cricket_t20_csv/t20_matchSummary.csv',index=False)
```

### 2nd : BATTING SUMMARY

```python
with open('cricket_t20_json_files/t20_wc_batting_summary.json') as f:
    data=json.load(f)
    
    all_records=[]
    
    for record in data:
        all_records.extend(record['battingSummary'])

all_records[0:2]
```

```python
df_bat = pd.DataFrame(all_records)

print(df_bat.shape)  # check no. of rows and columns
df_bat.head(10)
```

###### Creating a new column to check for the Out and Not-Out players

```python
df_bat['Out/Not-Out']=df_bat.dismissal.apply(lambda x :'Out' if len(x)>0 else 'Not-Out')
df_bat.head()
```

```python
# Removing 'Dismissal' column
df_bat.drop(columns=['dismissal'],inplace=True)
df_bat.head(10)
```

```python
# Taking care of errors in 'batsmanName' column

df_bat['batsmanName']=df_bat['batsmanName'].apply(lambda x: x.replace('â€',''))
df_bat['batsmanName']=df_bat['batsmanName'].apply(lambda x: x.replace('\xa0',''))
```

###  Mapping ' match_dict 'dictionary with ' battingSummary ' data

```python
df_bat["match_id"] = df_bat['match'].map(match_dict)

df_bat.head()
```

```python
# Exporting the file
df_bat.to_csv('cricket_t20_csv/t20_battingSummary.csv',index=False)
```
